```
# Package Used
import pandas as pd
import numpy as np
import re
```

```
train = pd.read_csv(r'/content/drive/Shared drives/RMSE就是我们的KPI/my.train1.csv')
```

```
test = pd.read_csv(r'/content/drive/Shared drives/RMSE就是我们的KPI/my.test1.csv')
```

```
all=pd.concat([train, test])
all=all.reset_index(drop=True)
all
```

```
all[['Full_ID','Timestamp']] = all.sessionId.str.split("_",expand=True,)
```

```
from pytz import country_timezones
```

```
all['date'] = pd.to_datetime(all['Timestamp'], unit='s',errors='coerce').dt.tz_localize('UTC')
```

```
Country_code={ 'India' : 'IN', 'United States':'US', 'Germany':'DE', 'Australia':'AU', 'Canada':'CA',
       'Japan':'JP', 'others':None, 'Russia':'RU', 'United Kingdom': 'GB', 'Brazil':'BR', 'Turkey':'TR',
       'Poland':'PL', 'Philippines':'PH', 'Netherlands': 'NL', 'Spain':'ES', 'Mexico':'MX',
       'Thailand':'TH', 'Taiwan':'CN', 'Vietnam':'VN', 'Indonesia':'ID', 'France':'FR', 'Italy':'IT'}
```

```
all['country_code'] = all['country'].replace(Country_code)
```

```

mask = all['country_code'].notnull() & (all['country_code'] != '')
```

```
mask.unique()
```

```

all['tz'] = all[mask].apply(lambda x: x['date'].tz_convert(country_timezones(x['country_code'])[0]), 
                          axis=1)
```

```
all['period'] = (all['tz'].dt.hour % 24 + 4) // 4
all['period'].replace({1: 'Late Night',
                      2: 'Early Morning',
                      3: 'Morning',
                      4: 'Noon',
                      5: 'Evening',
                      6: 'Night'}, inplace=True)
```
